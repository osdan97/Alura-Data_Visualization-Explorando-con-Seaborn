# Traduciendo variables

```
datos.columns
```

```
dic_columnas = {
    'LIMIT_BAL' : 'limite', 
    'CHECKING_ACCOUNT' : 'cuenta_corriente', 
    'EDUCATION' : 'escolaridad', 
    'MARRIAGE' : 'estado_civil', 
    'AGE' : 'edad',
    'BILL_AMT' : 'valor_factura', 
    'PAY_AMT' : 'valor_pago', 
    'DEFAULT' : 'moroso'
}
```

```
tarjetas = datos.rename(columns = dic_columnas)
```

```
tarjetas.head()
```

```
tarjetas.cuenta_corriente.unique()
```

```
dic_cuenta = {
    'Yes' : 'Si', 
    'No' : 'No'
}
```

```
tarjetas.cuenta_corriente.map(dic_cuenta)
```

```
tarjetas.cuenta_corriente = tarjetas.cuenta_corriente.map(dic_cuenta)
```

```
tarjetas.head()
```

```
tarjetas.escolaridad.unique()
```

```
dic_escolaridad = {
    '2.University' : '2.Universidad', 
    '3.Graduate School' : '3.Pos-graduaciÃ³n', 
    '1.High School' : '1.Colegio'
}
```

```
tarjetas.escolaridad = tarjetas.escolaridad.map(dic_escolaridad)
```

```
tarjetas.head()
```

```
from scipy.stats import ranksums
```

H<sup>null<sup>
La distibucion grafica de los morosos y no morosos es la misma

H<sup>alt<sup>

La distribucion grafica de los morosos y no morosos no es la misma

```
morosos=tarjetas.query('moroso==1').valor_factura
morosos
```

```
no_morosos=tarjetas.query('moroso==0').valor_factura
no_morosos
```

```
resultado=ranksums(morosos,no_morosos)
print('El pvalue de la hipotesis es {}'.format(resultado.pvalue))
```
